```python
# 라이브러리 추가
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
path = "./data"
file_name = 'total_car_info.csv'
target_file = os.path.join(path, file_name)
```

```python
df = pd.read_csv(target_file)
```

```python
# 친환경 차량 집계
tmp = df.groupby(['year', 'month', '친환경'])['계'].sum()
tmp = tmp.unstack()['친환경']
```

```python
tmp.plot()
```

```python
pd.Series([tmp[i + 1] - tmp[i] for i in range(len(tmp) - 1)]).plot()
```

```python
tmp = df[df['연료별'] == '수소'].groupby(['year_month', '친환경'])['계'].sum()
tmp = tmp.unstack()['친환경'].plot()
```
